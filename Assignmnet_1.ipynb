# Creating a DataFrame

```python
import pandas as pd

data = {
    "Brand": ["Samsung","Apple","Xiaomi","Oppo","Vivo","Realme","OnePlus","Nokia","Motorola","Pixel","Huawei"],
    "Model": [
        "S24 Ultra","iPhone 16 Pro","Xiaomi 14","Find X7"," X100 Pro","GT 6","OnePlus 12","Nokia X30","Edge 40","Pixel 8 Pro","Mate 50 pro"
    ],
    "Price": [20000,70000,15000,18000,17000,14000,35000,16000,19000,60000,20000],
    "Rating": ["NaN",4.8,"NaN",4.2,4.1,"NaN",4.6,3.9,"NaN",4.7,4.1]
}

df = pd.DataFrame(data)
df
```

**REPLACING THE "NaN"** WITH THE AVG VALUE FOR RATING

```python
c = avg =0;
for element in df["Rating"]:
    if element != "NaN":
        c += 1;
        avg += element;
avg = avg/c;

df = df.replace(to_replace="NaN", value= avg);

df
```

**DATA TYPE** FOR EACH COLUMN IN DATA FRAME

```python
df_cols = df.dtypes;
df_cols
```
