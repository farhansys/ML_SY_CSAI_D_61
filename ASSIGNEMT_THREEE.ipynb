# 1. SCATTER PLOT

```python
import pandas as pd

scatter_df = pd.DataFrame({
    "Study_Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Marks_Class_A": [30, 42, 45, 55, 60, 70, 68, 85],
    "Marks_Class_B": [28, 35, 40, 50, 58, 63, 75, 78]
})

print(scatter_df)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(
    scatter_df["Study_Hours"],scatter_df["Marks_Class_A"],color="orange",label="Class A"
)

plt.scatter(
    scatter_df["Study_Hours"],scatter_df["Marks_Class_B"],color="yellow",label="Class B"
)

plt.xlabel("Study Hours", fontsize=12)
plt.ylabel("Marks Obtained", fontsize=12)
plt.title("Scatter Plot Showing Correlation Between Study Hours and Marks", fontsize=14)

plt.legend(title="Class")
plt.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

# 2. BAR CHART


```python
bar_df = pd.DataFrame({
    "Product": ["Product A", "Product B", "Product C", "Product D"],
    "Sales_2024": [120, 150, 100, 170],
    "Sales_2025": [140, 160, 130, 180]
})

print(bar_df)
```

```python
x = np.arange(len(bar_df["Product"]))
print(x)
width = 0.35

plt.figure(figsize=(10, 6))

plt.bar(x - width/2, bar_df["Sales_2024"], width, label="Sales 2024")
plt.bar(x + width/2, bar_df["Sales_2025"], width, label="Sales 2025")

plt.xlabel("Products", fontsize=12)
plt.ylabel("Sales (Units)", fontsize=12)
plt.title("Bar Chart: Product Sales Comparison (2024 vs 2025)", fontsize=14)

plt.xticks(x, bar_df["Product"])

plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

# 3. BOX PLOT

```python
box_df = pd.DataFrame({
    "Salary": [
        12000,28000, 30000, 32000, 34000, 36000, 38000, 40000, 42000, 45000, 48000,60000
    ]
})

```

# 4. PIE CHART

```python
pie_df = pd.DataFrame({
    "Department": ["IT", "HR", "Sales", "Finance"],
    "Employees": [40, 20, 25, 15]
})
```

```python
plt.figure(figsize=(8, 7))

plt.pie(
    pie_df["Employees"],
    labels=pie_df["Department"],
    autopct="%1.1f%%",
    startangle=180
)

plt.title("Pie Chart Showing Department-wise Employee Distribution", fontsize=14)

plt.legend(
    pie_df["Department"],
    title="Departments",
    loc="best"
)

plt.show()
```

# 5. LINE GRAPH

```python
line_df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [120, 135, 150, 165, 180, 200],
    "Profit": [30, 38, 45, 55, 65, 75],
    "Expenses": [90, 97, 105, 110, 115, 125]
})
```

```python
plt.figure(figsize=(10, 6))

plt.plot(line_df["Month"], line_df["Sales"],marker="o",linewidth=2, label="Sales")

plt.plot(line_df["Month"], line_df["Profit"],marker="o", linewidth=2, label="Profit")

plt.plot(line_df["Month"], line_df["Expenses"],marker="o", linewidth=2, label="Expenses")

plt.xlabel("Month", fontsize=12)
plt.ylabel("Amount (in units)", fontsize=12)
plt.title("Line Graph Showing Monthly Sales, Profit, and Expenses", fontsize=14)

plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

# 6. HEAT MAP

```python
heatmap_df = pd.DataFrame({
    "Maths": [78, 85, 88],
    "Science": [72, 80, 90],
    "English": [68, 75, 82],
    "Computer": [85, 90, 95]
}, index=["Midterm", "Internal", "Final"])
```

```python
import seaborn as sns

plt.figure(figsize=(9, 6))

ax = sns.heatmap(
    heatmap_df,
    annot=True,
    fmt="d",
    cmap="YlOrRd",
    linewidths=0.5,
    cbar=True
)

plt.title("Heat Map Showing Student Performance Across Subjects and Exams", fontsize=14)
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Exam Type", fontsize=12)

cbar = ax.collections[0].colorbar
cbar.set_label("Marks", fontsize=11)

plt.show()
```
