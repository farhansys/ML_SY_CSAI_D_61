Creating User Define Functions for the Built-in functions

```python
import pandas as pd

numbers = [12, 15, 12, 18, 20, 15, 25]

def mean(numbers):
    return sum(numbers) / len(numbers)

def minimum(numbers):
    return min(numbers)

def maximum(numbers):
    return max(numbers)

def median(numbers):
    sorted_nums = sorted(numbers)
    n = len(sorted_nums)
    if n % 2 == 0:
        return (sorted_nums[n//2 - 1] + sorted_nums[n//2]) / 2
    else:
        return sorted_nums[n//2]

def mode(numbers):
    freq = {}
    for num in numbers:
        freq[num] = freq.get(num, 0) + 1
    max_freq = max(freq.values())
    return [k for k, v in freq.items() if v == max_freq]

def variance(numbers):
    m = mean(numbers)
    return sum((x - m) ** 2 for x in numbers) / (len(numbers) - 1)

def std(numbers):
    return variance(numbers) ** 0.5

def range(numbers):
    return maximum(numbers) - minimum(numbers)

results = {
    "Mean": mean(numbers),
    "Minimum": minimum(numbers),
    "Maximum": maximum(numbers),
    "Median": median(numbers),
    "Mode": mode(numbers),
    "Variance": variance(numbers),
    "Standard Deviation": std(numbers),
    "Range": range(numbers)
}

df = pd.DataFrame(list(results.items()), columns=["Statistic", "Value"])
df
```

# CHECKPOINT 3: DATA INTEGRATION

Vertical Data Integration (Row-wise)
Combines datasets with same structure by stacking rows.

```python
import pandas as pd

jan = pd.DataFrame({
    'EmpID': [1, 2],
    'Sales': [20000, 25000]
})

feb = pd.DataFrame({
    'EmpID': [3, 4],
    'Sales': [22000, 27000]
})

sales = pd.concat([jan, feb], ignore_index=True)

print(sales)
```

```python
marks = pd.DataFrame({
    'RollNo': [101, 102, 103],
    'Marks': [85, 90, 88]
})

attendance = pd.DataFrame({
    'RollNo': [101, 102, 103],
    'Attendance': [92, 95, 90]
})

student_data = pd.concat([marks, attendance['Attendance']], axis=1)

print(student_data)
```

```python
students = pd.DataFrame({
    'StudentID': [1, 2, 3],
    'Name': ['Amit', 'Neha', 'Rahul']
})

results = pd.DataFrame({
    'StudentID': [1, 2],
    'CGPA': [8.5, 9.1]
})

data_inner = pd.merge(students, results, on='StudentID', how='inner')

print(data_inner)
```

Types of Joins in Data Integration

```python
pd.merge(students, results, on='StudentID', how='inner')
pd.merge(students, results, on='StudentID', how='left')
pd.merge(students, results, on='StudentID', how='right')
pd.merge(students, results, on='StudentID', how='outer')
```

Integration Using Index Using join() Joins datasets using index instead of columns.

```python
df1 = students.set_index('StudentID')
df2 = results.set_index('StudentID')

integrated_data = df1.join(df2)

print(integrated_data)
```

# CHECKPOINT 4: DATA TRANSFORMATION

1. MinMaxScaler vs StandardScaler

Both are feature scaling techniques, but they work differently.

A. MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

df = pd.DataFrame({'Age':[18,22,25,30,35]})
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
print(df_scaled)
```

A. Scaling / Normalization / Standardization

Purpose: Make numerical features comparable.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd

df = pd.DataFrame({'Age':[18,22,25,30], 'Salary':[25000,50000,60000,80000]})

scaler = MinMaxScaler()
df[['Age','Salary']] = scaler.fit_transform(df[['Age','Salary']])
print("Min-Max Scaled:\n", df)

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['Age','Salary']])
print("Standardized:\n", df_scaled)
```
